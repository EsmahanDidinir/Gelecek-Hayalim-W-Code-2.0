```python

# Veri setini oku
df = pd.read_csv("Iris.csv")

# Türlere göre ortalama değerler
mean_values = df.groupby("Species")[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]].mean()

# Çubuk grafik
mean_values.plot(kind="bar", figsize=(8,6))
plt.title("Çiçek Türlerine Göre Ortalama Ölçüler")
plt.ylabel("Ortalama Uzunluk/Genişlik (cm)")
plt.xlabel("Çiçek Türü")
plt.xticks(rotation=0)
plt.legend(title="Özellik")
plt.show()
```

```python
import seaborn as sns

# Scatter plot (dağılım grafiği)
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="PetalLengthCm",
    y="PetalWidthCm",
    hue="Species",   # renklendirme türlere göre
    style="Species", # şekil türlere göre
    s=80             # nokta boyutu
)
plt.title("Türlere Göre Petal Uzunluğu ve Genişliği")
plt.show()
```
