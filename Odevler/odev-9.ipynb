```python

# Veri setini oku
df = pd.read_csv("Iris.csv")

# Trlere gre ortalama deerler
mean_values = df.groupby("Species")[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]].mean()

# ubuk grafik
mean_values.plot(kind="bar", figsize=(8,6))
plt.title("iek Trlerine Gre Ortalama ller")
plt.ylabel("Ortalama Uzunluk/Genilik (cm)")
plt.xlabel("iek Tr")
plt.xticks(rotation=0)
plt.legend(title="zellik")
plt.show()
```

```python
import seaborn as sns

# Scatter plot (dalm grafii)
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="PetalLengthCm",
    y="PetalWidthCm",
    hue="Species",   # renklendirme trlere gre
    style="Species", # ekil trlere gre
    s=80             # nokta boyutu
)
plt.title("Trlere Gre Petal Uzunluu ve Genilii")
plt.show()
```
